## 과제 개요

삼성전자의 2014–2024년 감사보고서 HTML 파일을 활용하여 금융 도메인 특화 NLP 시스템을 구축한다.

HTML 파싱과 기본적인 데이터 파이프라인 구성을 토대로, CI/CD 환경을 갖춘 **완성도 있는 최종 응용 시스템**을 개발하는 것이 목표이다.

---

## 과제 목표

10년치 삼성전자 감사보고서를 처리하는 **데이터 파이프라인**과, 이를 기반으로 동작하는 **금융 NLP 응용 시스템**을 E2E로 구현한다.

```
[HTML 감사보고서 10개년 (2014–2024)]
        │
        ▼  HTML 파싱 & 텍스트 정제
        │
        ▼  기본 데이터 스키마 설계 (CSV / JSON / SQLite)
        │
        ▼  CI/CD 파이프라인 구성
        │
        ▼  최종 응용 시스템
```

---

## 구현 단계

### Step 1. HTML 파싱 및 데이터 전처리

2–3회차 실습을 기반으로, 10개년 감사보고서에 **일반화 가능한 파싱 모듈**을 개발한다.

- **HTML 구조 분석**: 연도별 구조적 차이를 파악하고 공통 파싱 패턴을 도출한다.
- **섹션 식별**: 재무제표, 주석, 감사의견 등 주요 섹션을 자동 분류한다.
- **표 데이터 추출**: HTML 테이블을 구조화된 `DataFrame`으로 변환한다.


---

### Step 2. 기본 데이터 스키마 설계

10개년 자료를 일관되게 담을 수 있는 **최소한의 통일된 스키마**를 설계한다.  
복잡한 DB 인프라 구축보다는 **재현 가능한 구조**를 우선으로 한다.

- **스키마 설계**: 연도·섹션·항목을 키로 하는 통일된 데이터 구조를 정의한다.
- **저장 형식**: `CSV` / `JSON` / `SQLite` 중 하나를 선택한다. (복잡한 DB 인프라는 불필요)
- **메타데이터 포함**: 소스 파일명, 추출 일시, 연도 등 기본 메타정보를 함께 저장한다.
- **데이터 검증**: 누락·이상값 탐지 로직과 처리 전략을 코드 또는 문서로 명시한다.

---

### Step 3. CI/CD 파이프라인 구성

코드 변경이 자동으로 검증되는 **체계적인 개발 환경**을 구성한다.

- **버전 관리**: Git 브랜치 전략 및 커밋 컨벤션을 설정한다.
- (선택)**자동화 테스트**: 파싱 모듈·전처리 함수에 대한 단위 테스트를 작성한다. (`pytest`)
- (선택)**CI 워크플로우**: GitHub Actions 기반 자동 테스트 실행을 구성한다. (`.github/workflows/`)
- **환경 재현성**: `requirements.txt` 또는 `pyproject.toml`로 의존성을 고정한다. (Docker는 선택)

---

### Step 4. 최종 응용 시스템 개발: RAG 기반 QA 모델

RAG·스트리밍·Tool Calling 파이프라인을 **10개년 감사보고서 전체로 확장**하여, 연도 간 비교 질의가 가능한 금융 QA 시스템을 완성한다. (Orchestration 기반 에이전틱 워크플로우 허용)

```
[전처리된 10개년 감사보고서 (Step 1–2 산출물)]
        │
        ▼  청킹 & 임베딩 (sentence-transformers)
        │
        ▼  벡터 DB 구축 (FAISS / ChromaDB)
        │
        ▼  하이브리드 검색 (의미 검색 + 키워드 필터)
        │
        ▼  LLM 기반 답변 생성 (RAG + Tool Calling)
        │
        ▼  멀티턴 QA 인터페이스 (Streamlit / CLI)
```
